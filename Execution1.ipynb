```python
#!/usr/bin/env python
# coding: utf-8



# # The top two Companies for cloud services providers
#
#
# **AWS** One of the major providers for storage and computing in the cloud
#
# **Microsoft** The oldest Tech company provide a wide range of services including cloud computing
#
import matplotlib.pyplot as plt
import numpy as np
get_ipython().run_line_magic('matplotlib', 'inline')
x=[2,4,6,8,10,12,14,15,16]
y1=[2,3,2,3,5,7,2,3,8]
y2=[1,2,5,4,6,8,6,7,3]
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_title('Comparative Graph',fontsize=24)



ax.set_xlabel('-----x axis-----')
ax.set_ylabel('-----y axis-----')




plt.plot(x,y1,label='AWS')
plt.plot(x,y2,label='Microsoft')

ax.legend()

print("Graph of Progression Timeline")

```
